```python
sector_values = data.sector.value_counts().to_dict()
sector_values
```

```python
sector_list = []
for key, value in sector_values.items():
    if value > 1000:
        sector_list.append(key)
len(sector_list)
```

```python
job_values = data.job.value_counts().to_dict()
```

```python
job_values
```

```python
job_list = []
for key, value in job_values.items():
    if value > 1200:
        job_list.append(key)
len(job_list)
```

```python
from sklearn.preprocessing import OneHotEncoder
def categorical_encoding(column_name=None, data_frame=None):
    wclass_encoder = OneHotEncoder()
    ohe = wclass_encoder.fit_transform(data_frame[column_name].values.reshape(-1,1)).toarray()
    dfOneHot = pd.DataFrame(ohe, columns = [column_name+"_"+str(wclass_encoder.categories_[0][i]) for i in range(len(wclass_encoder.categories_[0]))])
    data_frame = pd.concat([data_frame, dfOneHot], axis=1)
    data_frame = data_frame.drop(column_name, axis=1)
    return data_frame
```

```python
data['sector'] = ['other' if x not in sector_list else x for x in data['sector']]
data.head()
```

```python
data.sector.unique()
```

```python
data['job'] = ['other' if x not in job_list else x for x in data['job']]
data.head()
```

```python
data.job.nunique()
```

```python
data = categorical_encoding(column_name='job',data_frame=data)
data.head()
```

```python
data = categorical_encoding(column_name='sector', data_frame=data)
data.head()
```
