```python
import pandas as pd
import matplotlib as plt
```

```python
census_data = pd.read_csv(filepath_or_buffer='census_no_null.csv')
census_data.head()
```

```python
census_data.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
def categorical_encoding(column_name=None, data_frame=None):
    wclass_encoder = OneHotEncoder()
    ohe = wclass_encoder.fit_transform(data_frame[column_name].values.reshape(-1,1)).toarray()
    dfOneHot = pd.DataFrame(ohe, columns = [column_name+"_"+str(wclass_encoder.categories_[0][i]) for i in range(len(wclass_encoder.categories_[0]))])
    data_frame = pd.concat([data_frame, dfOneHot], axis=1)
    data_frame = data_frame.drop(column_name, axis=1)
    return data_frame
```

```python
census_data = categorical_encoding(column_name='workclass', data_frame=census_data)
census_data.head()
```

```python
census_data = categorical_encoding(column_name='marital_status', data_frame=census_data)
census_data.head()
```

```python
census_data = categorical_encoding(column_name='occupation', data_frame=census_data)
census_data.head()
```

```python
census_data = categorical_encoding(column_name='relationship', data_frame=census_data)
census_data.head()
```

```python
census_data = categorical_encoding(column_name='race', data_frame=census_data)
census_data.head()
```

```python
census_data = categorical_encoding(column_name='sex', data_frame=census_data)
census_data.head()
```

```python
census_data = categorical_encoding(column_name='native_country', data_frame=census_data)
census_data.head()
```
